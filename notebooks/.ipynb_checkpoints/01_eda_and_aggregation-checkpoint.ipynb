```python
# Parse timestamps with mixed formats
df["timeStamp"] = pd.to_datetime(
    df["timeStamp"],
    format="mixed",
    dayfirst=True
)
```

```python
df["timeStamp"].dtype
```

```python
# Use the timestamp as the index for time-series operations
df = df.set_index("timeStamp")

df.head()
```

```python
# Aggregate total number of 911 calls per day
daily_calls = df.resample("D").size()

daily_calls.head()
```

```python
 # Make sure every day exists in the time series
daily_calls = daily_calls.asfreq("D", fill_value=0)

daily_calls.isna().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(daily_calls)
plt.title("Daily 911 Call Volume (Philadelphia Metro Area)")
plt.xlabel("Date")
plt.ylabel("Number of Calls")
plt.tight_layout()
plt.show()
```

```python
# Average call volume by day of week
weekday_avg = daily_calls.groupby(daily_calls.index.dayofweek).mean()

weekday_avg
```

```python
plt.figure(figsize=(8, 4))
weekday_avg.plot(kind="bar")
plt.title("Average 911 Calls by Day of Week")
plt.xlabel("Day of Week (0=Mon, 6=Sun)")
plt.ylabel("Average Calls")
plt.tight_layout()
plt.show()
```

```python
weekday_labels = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt.figure(figsize=(8, 4))
weekday_avg.plot(kind="bar")
plt.xticks(range(7), weekday_labels, rotation=0)
plt.title("Average 911 Calls by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Average Calls")
plt.tight_layout()
plt.show()
```

```python
# Average call volume by month
monthly_avg = daily_calls.groupby(daily_calls.index.month).mean()

monthly_avg
```

```python
import matplotlib.pyplot as plt

month_labels = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

plt.figure(figsize=(8, 4))
monthly_avg.plot(kind="bar")
plt.xticks(range(1, 13), month_labels, rotation=0)
plt.title("Average 911 Calls by Month")
plt.xlabel("Month")
plt.ylabel("Average Calls")
plt.tight_layout()
plt.show()
```

### EDA Summary
